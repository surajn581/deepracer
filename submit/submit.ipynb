```python
def open_browser():
    options = Options()
    # options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    # options.add_argument("--single-process")
    # options.add_argument("--disable-dev-shm-usage")
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)"
    )

    # browser = webdriver.Chrome(options=options)
    # browser = webdriver.Edge(r"msedgedriver.exe")
    browser = webdriver.Edge(options = options)

    # browser = webdriver.Chrome("/usr/local/bin/chromedriver")

    browser.set_window_size(1600, 1700)

    return browser
```

```python
def scroll_to_view( element ):
    while not ( element.is_displayed() and element.is_enabled() ):
        time.sleep(0.5)
    element.location_once_scrolled_into_view
    
def doSubmit(browser, model_name):
    # url = r"https://us-east-1.console.aws.amazon.com/deepracer/home?region=us-east-1#competition/arn%3Aaws%3Adeepracer%3A%3A483111947419%3Aleaderboard%2F1ce3185b-e020-4307-8bcd-29bfce9f08d7/submitModel"
    url = r"https://us-east-1.console.aws.amazon.com/deepracer/home#competition/arn%3Aaws%3Adeepracer%3A%3A091259800912%3Aleaderboard%2F9610c639-e8d4-455b-8879-f4744d98fc8a/submitModel"
    browser.get(url)
    browser.refresh()
    delay = 30
    # dropdown = WebDriverWait(browser, delay).until(EC.presence_of_element_located((By.XPATH, '//span[text()="Choose a model"]')))
    dropdown = WebDriverWait(browser, delay).until(EC.element_to_be_clickable((By.XPATH, '//span[text()="Choose a model"]')))
    scroll_to_view( dropdown )
    dropdown.click()

    # modelbutton = browser.find_element(By.XPATH, '//span[text()="{}"]'.format(model_name))
    modelbutton = WebDriverWait(browser, delay).until(EC.element_to_be_clickable((By.XPATH, '//span[text()="{}"]'.format(model_name))))
    scroll_to_view(modelbutton)
    modelbutton.click()

    # enter_race_button = browser.find_element(By.XPATH, '//span[text()="Enter race"]')
    enter_race_button = WebDriverWait(browser, delay).until(EC.element_to_be_clickable((By.XPATH, '//span[text()="Enter race"]')))
    scroll_to_view(enter_race_button)
    enter_race_button.click()
```

```python
model_names = ['MUDR24-219-Model-SN-250', 'MUDR24-219-Model-SN-360p1']
```

```python
def run():
    MINS_TO_SLEEP = 12
    while True:
        for model_name in model_names:
            try:
                print( datetime.datetime.now(), ' Submitting Model: ', model_name)
                doSubmit(browser, model_name)
                print( datetime.datetime.now(), ' Submission Successful, next submission will be triggered in ', MINS_TO_SLEEP, ' minutes')
                time.sleep(60*MINS_TO_SLEEP)
            except Exception as ex:
                print( datetime.datetime.now(), ' Could not submit due to: ', ex)
                print( datetime.datetime.now(), ' Resubmission will be triggered in 5 seconds')
                time.sleep(5)
            print('-'*100)
```

```python
run()
```
